```python
while True:
        print("\nCalculator Menu:")
        print("1: Sum Numbers")
        print("2: Multiplication")
        print("3: Division")
        print("4: Reciprocal (1/x)")
        print("5: Exponentiation (Base Power)")
        print("6: Factorial")
        print("0: Exit")

        choice = input("Enter your choice: ")

        if choice == "0":
                print("Exiting Calculator\nGood Bye!")
                break
        

        elif choice == "1":
                #Sum Numbers
                print("\nDoing Sum")
                num_to_add = input("Enter the numbers to add like: 1,-3,24,2,..: ") #taking the data in comma saperated string
                numbers = num_to_add.split(",")  #splitting the numbers by commas

                numbers_to_show = num_to_add.replace(",", " + ")  #replacing commas to add just to better visualisation

                total = 0    # initization total to 0
                for num in numbers:
                    total += float(num)
                print(f"The total of {numbers_to_show} = {total}")


        elif choice == "2":
                #Multiplication
                print("\nDoing Multiplication")
                num_to_multi = input("Enter the numbers to multiply like: 1,-3,24,2,..: ") #taking the data in comma saperated string
                numbers = num_to_multi.split(",")  #splitting the numbers by commas

                numbers_to_multi = num_to_multi.replace(",", " x ")  #replacing , to x just for better visualisation

                total = 1.0    # initization total to 1
                for num in numbers:
                    total *= float(num)
                print(f"The total of {numbers_to_multi} = {total}")



        elif choice == "3":
                #Division
                print("\nDoing Division")
                dividend = float(input("Enter Dividend (number you want to divide): "))
                divisor = float(input("Enter Divisor (number you want to divide with): "))

                if(divisor == 0):       #if divisor is zero error will occur 
                    print("Undefined\nDividing any number by zero does not make sense")
                else:
                    result = dividend/divisor   #finding the result

                    quotient = dividend//divisor    #for getting only the quotient

                    remainder = dividend%divisor    #for getting only the remainder

                    print(f"{dividend} / {divisor} = {result}  \nQuotient = {quotient}\nRemainder = {remainder}")   #printing everything



        elif choice == "4":
                #Reciprocal (1/x)
                print("\nDoing Reciprocal")
                number_to_reciprocal = float(input("Enter a number to get it's reciprocal: "))

                if number_to_reciprocal == 0:   #avoiding division by zero
                    print(f"The reciprocal of {number_to_reciprocal} is Undefined")
                else:
                    reciprocal = 1/number_to_reciprocal         #finding the reciprocal
                    print(f"The reciprocal of {number_to_reciprocal} is {reciprocal}")



        elif choice == "5":
                #Exponentiation (Base Power)
                print("Doing Exponentiation")
                base = float(input("Enter the Base number: "))
                power = float(input("Enter the Power number: "))

                result = base**power    
                print(f"{base} to the power {power} is {result}")



        elif choice == "6":
                #Factorial
                print("\nFactorial")
                number_for_factorial = int(input("Enter a number to get its factorial: "))

                if number_for_factorial<0:
                    print("Please enter a positive integer")
                else:
                    factorial = 1
                    for i in range(1, number_for_factorial+1):
                        factorial *= i
                    print(factorial)



        else:
                 print("Invalid choice. Please try again.")
```
